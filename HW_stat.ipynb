### surgery?

Может принимать только значение 1 или 2. Заменим на значение моды

```python
df['surgery?'].fillna(df['surgery?'].mode()[0], inplace=True)
```

### temperature of extremities, pain
это субьективные оценки и сложно установить взаимосвять с другими показателями
### outcome
так же может принимать разные значения при равных условиях, поэтому заполним пропуски последующими значениями df

```python
df.pain.fillna(method='ffill', inplace=True)
df['temperature of extremities'].fillna(method='ffill', inplace=True)
df.outcome.fillna(method='ffill', inplace=True)
```

### rectal temperature, respiratory rate, pulse

чтобы рассчитать данные показатели, добавим еще столбцы с округленными значениями до целого для температуры и до десятков для частоты дыхания и пульса. Таким образом мы сможем сгруппировать значения. 

```python
df['pulse_round'] = (df['pulse']/10).round()*10
df['respiratory_rate_round'] = (df['respiratory rate']/10).round()*10
df['rectal_temperature_round'] = (df['rectal temperature']).round()
```

```python
# частота дыхания повышается с учащением пульса
df.groupby('pulse_round')['respiratory rate'].median()
```

```python
# пульс учащается с отклонением от нормы
df.groupby('rectal_temperature_round')['pulse'].median()
```

```python
# температура повышается и дыхание учащается
df.groupby('respiratory_rate_round')['rectal temperature'].mean()
```

```python
df['respiratory rate'].fillna(df.groupby('pulse_round')['respiratory rate'].transform('median'), inplace=True)
df['pulse'].fillna(df.groupby('rectal_temperature_round')['pulse'].transform('median'), inplace=True)
df['rectal temperature'].fillna(df.groupby('respiratory_rate_round')['rectal temperature'].transform('mean').round(1), inplace=True)
```

```python
pd.concat([df.describe(), df.mode()])
```

```python
# все еще есть пропуски, но их уже немного. Заменим на значение медианы

df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace=True)
df['pulse'].fillna(df['pulse'].median(), inplace=True)
df['rectal temperature'].fillna(df['rectal temperature'].median(), inplace=True)
```
